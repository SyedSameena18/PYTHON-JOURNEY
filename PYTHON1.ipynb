```python
import random
p1 = input("Enter player1: ")
p2 = input("Enter player2: ")
s1 = 10
s2 = 20
d1 = random.randint(1, 10)
d2 = random.randint(1, 10)
print("_____PLAYER1 TURN_____")
while True:
    g1 = int(input("Enter your guess: "))
    s1 = s1 - 1
    if g1 == d1:
        print("Correct guess!")
        break
print("_____PLAYER2 TURN_____")
while True:
    g2 = int(input("Enter your guess: "))
    s2 = s2 - 1
    if g2 == d2:
        print("Correct guess!")
        break
print("________SUMMARY________")
print("{} : {}".format(p1, s1))
print("{} : {}".format(p2, s2))
if s1 > s2:
    print(p1, "wins")
elif s2 > s1:
    print(p2, "wins")
else:
    print("Draw")

```
